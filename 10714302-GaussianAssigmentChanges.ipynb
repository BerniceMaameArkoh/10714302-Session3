```python
import numpy as np
from math import pi, e
import matplotlib.pyplot as plt 
import cv2 
```

```python
# Gaussian kernel generation
def gauss_kernel(kernel_size=3, sigma=1):
    kernel = np.fromfunction(lambda x, y: (1/(2*pi*sigma**2)) * e ** ((-1*((x-(kernel_size-1)/2)**2+(y-(kernel_size-1)/2)**2))/(2*sigma**2)), (kernel_size, kernel_size))
    return kernel / np.sum(kernel)
```

```python
gauss_kernel()
```

```python
def gauss_blur(img, Kernel_size, Sigma):

    kernel = gauss_kernel(Kernel_size, Sigma)
    d = int((Kernel_size-1)/2)
    Gaussian = np.zeros((img.shape[0]-2*d, img.shape[1]-2*d))
    for y in range(d, img.shape[0]-d):
        for x in range(d, img.shape[1]-d):
            Gaussian[y-d][x-d] = np.sum(np.multiply(img[y-d:y+d+1, x-d:x+d+1], kernel))
    return Gaussian
```
